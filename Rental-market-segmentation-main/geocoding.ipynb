```python
airport_loc = geolocator.geocode('Pune international airport')
airport_latitude = airport_loc.latitude
airport_longitude = airport_loc.longitude
```

```python
railway_loc = geolocator.geocode('Pune railway station')
railway_lat = railway_loc.latitude
railway_lng = railway_loc.longitude
```

```python
import geopy.distance
```

```python
def get_distance_airport(lat_in,lng_in):

  coords_1 = (lat_in,lng_in)
  coords_2 = (airport_latitude,airport_longitude)
  
  return geopy.distance.vincenty(coords_1, coords_2).km
```

```python
def get_distance_railway(lat_in,lng_in):

  coords_1 = (lat_in,lng_in)
  coords_2 = (railway_lat,railway_lng)
  
  return geopy.distance.vincenty(coords_1, coords_2).km
```

```python
d_a = []
d_r = []
```

```python
d_a
```

```python

for i in add1:
  ad = get_distance_airport(alat[i],alng[i])
  rd = get_distance_railway(alat[i],alng[i])
  if ad<100 and rd<100:
    d_a.append(ad)
    d_r.append(rd)
  else:
    d_a.append(None)
    d_r.append(None)
```

```python
df['da'] = d_a
df['dr'] = d_r
```
